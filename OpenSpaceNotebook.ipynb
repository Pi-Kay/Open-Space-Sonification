```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from midiutil import MIDIFile
import re

# Your custom str2midi (kept from before)
def str2midi(note):
    match = re.match(r'([A-Ga-g])(#|b)?(-?\d+)', note)
    if not match:
        raise ValueError(f"Invalid note: {note}")
    letter, accidental, octave_str = match.groups()
    letter = letter.upper()
    octave = int(octave_str)
    semitones = {'C': 0, 'D': 2, 'E': 4, 'F': 5, 'G': 7, 'A': 9, 'B': 11}
    semitone = semitones[letter]
    if accidental == '#': 
        semitone += 1
    elif accidental == 'b': 
        semitone -= 1
    return (octave + 1) * 12 + semitone

# Same rich note set as reference (Cmaj13#11 voicing)
note_names = ['C1','C2','G2','C3','E3','G3','A3','B3','D4','E4','G4','A4','B4','D5','E5','G5','A5','B5','D6','E6','F#6','G6','A6']
note_midis = [str2midi(n) for n in note_names]
n_notes = len(note_midis)

def map_value(value, min_value, max_value, min_result, max_result):
    return min_result + (value - min_value) / (max_value - min_value) * (max_result - min_result)

# Your placeholder data (10 craters) - keep for now, or replace with real later
df = pd.DataFrame({
    'age_myr': [3800, 3200, 1000, 800, 3800, 3200, 1000, 800, 3800, 3200],
    'diameter_km': [1160, 93, 100, 20, 250, 50, 300, 10, 500, 150],
    'intensity': np.random.rand(10)  # Not used here, like reference
})

n_impacts = len(df)
ages = df['age_myr'].values
times_myr = max(ages) - ages  # Oldest first, like reference
duration_beats = 60  # ~1 minute at 60 BPM, adjustable
t_data = map_value(times_myr, 0, max(times_myr), 0, duration_beats)
bpm = 60  # 1 beat = 1 second, like reference

# Normalize diameter (non-linear, like reference y_scale=0.5)
diameter_data = df['diameter_km'].values ** 0.5
diameter_norm = map_value(diameter_data, min(diameter_data), max(diameter_data), 0, 1)

# Map to MIDI: deeper/larger  lower note (inverse, like reference)
midi_data = []
for i in range(n_impacts):
    note_index = round(map_value(diameter_norm[i], 0, 1, n_notes-1, 0))  # Large  low index (low pitch)
    midi_data.append(note_midis[note_index])

# Velocity: larger  louder (like reference)
vel_min, vel_max = 35, 127
vel_data = [round(map_value(diameter_norm[i], 0, 1, vel_min, vel_max)) for i in range(n_impacts)]

# Create MIDI (single notes, fixed duration like reference)
my_midi_file = MIDIFile(1)
my_midi_file.addTempo(track=0, time=0, tempo=bpm)

for i in range(n_impacts):
    my_midi_file.addNote(track=0, channel=0, pitch=midi_data[i], time=t_data[i], duration=2, volume=vel_data[i])

filename = 'moon_crater_reference_style.mid'
with open(filename, 'wb') as f:
    my_midi_file.writeFile(f)

print(f"MIDI saved as '{filename}' - should have clear sound now!")

# Plot like reference
plt.scatter(t_data, midi_data, s=50 * diameter_norm)
plt.xlabel('Time [beats]')
plt.ylabel('MIDI Note Number')
plt.title('Moon Craters  Music (Reference Style)')
plt.show()
```
